**Lets analyze data**

```python
import pandas as pd
```

```python
data = pd.read_csv("all_perth_310121.csv")
data.head(10)
```

```python
data.tail(10)
```

```python
data.shape
```

```python
data.sample(10)
```

Deleting some unnecessary columns. I think unnecessary columns are " DATE_SOLD (house was last sold), POSTCODE, LATITUDE(Latitude of address), LONGITUDE(Longitude of address)"

```python
data_editing = data.drop(['DATE_SOLD', 'POSTCODE', 'LATITUDE', 'LONGITUDE'], axis=1)
data_editing.head(10)
```

```python
data_editing.shape
```

Deleting NaN rows

```python
data_editing_ = data_editing.dropna()
data_editing_.shape
```

```python
data_editing_.head(10)
```

```python
data_editing_['SUBURB'].drop_duplicates()
```

```python
print( 'Houses from ', len(data_editing_['SUBURB'].drop_duplicates()), ' Perth suburbs')
```

**Lets take Bayswater SUBURB**

```python
data_Bayswater = data_editing_.groupby('SUBURB').get_group('Bayswater')
data_Bayswater
```

```python
data_Bayswater_ = data_Bayswater.sort_values(by='PRICE', ascending=False).reset_index(drop=True)
data_Bayswater_
```

**Price of House in Bayswater from  260,000 to 940,000**

```python
# sorting by BUILD_YEAR and NEAREST SCHOOL  
data_Bayswater_1 = data_Bayswater.sort_values(by=['BUILD_YEAR', 'NEAREST_SCH_DIST'], ascending=[False, True]).reset_index(drop=True)
data_Bayswater_1
```

**Let's go back to our main data**

```python
print('Average Price: ', round(data_editing_['PRICE'].mean()))
print('Max Price: ', round(data_editing_['PRICE'].max()))
print('Min Price: ', round(data_editing_['PRICE'].min()), '\n')

print('Average land area: ', round(data_editing_['LAND_AREA'].mean()))
print('Max land area: ', round(data_editing_['LAND_AREA'].max()))
print('Min land area: ', round(data_editing_['LAND_AREA'].min()), '\n')


print('Average floor area: ', round(data_editing_['FLOOR_AREA'].mean()))
print('Max floor area: ', round(data_editing_['FLOOR_AREA'].max()))
print('Min floor area: ', round(data_editing_['FLOOR_AREA'].min()))
```

```python
data_editing_.groupby('PRICE').get_group(data_editing_['PRICE'].max())
```

```python
data_editing_.groupby('FLOOR_AREA').get_group(data_editing_['FLOOR_AREA'].max())
```

```python
# Price per square meter
data_editing_copy = data_editing_.copy()
data_editing_copy['price_per_sqm'] = data_editing_['PRICE'] / data_editing_['FLOOR_AREA']

data_editing_copy['price_per_sqm_land_area'] = data_editing_['PRICE'] / data_editing_['LAND_AREA']

# Square meters per rooms
data_editing_copy['sqm_per_room'] = data_editing_['FLOOR_AREA'] / (data_editing_['BEDROOMS'] + data_editing_['BATHROOMS'])

data_editing_copy
```

```python
print('Average Distance to Central Business District: ', round(data_editing_['CBD_DIST'].mean()))
print('Max Distance to Central Business District: ', round(data_editing_['CBD_DIST'].max()))
print('Min Distance to Central Business District: ', round(data_editing_['CBD_DIST'].min()), '\n')

print('Lets Find Distance to Central Business District atleast 2000 metres')
```

```python
out_sq = data_editing_copy[data_editing_copy['CBD_DIST'] < 2000]
out_sq
```

### Visualization

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
lol = pd.DataFrame(data_editing_copy['SUBURB'])
vis = lol.groupby(['SUBURB']).size().reset_index(name='Count')
vis
```

**Counting house in 276 SUBURBs. And visualization first 35 SUBURBs and last 26 SUBURBs**

```python
plt.figure(figsize=(15,10))
vis = vis.sort_values(by='Count',ascending=False)
sns.barplot(x=vis.SUBURB[:35],y=vis.Count[:35])
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(15,10))
vis = vis.sort_values(by='Count',ascending=False)
sns.barplot(x=vis.SUBURB[250:],y=vis.Count[250:])
plt.xticks(rotation=90)
plt.show()
```

```python
avg_price_by_SUBURB = data_editing_copy.groupby('SUBURB')['PRICE'].max().reset_index()
avg_price_by_SUBURB
```

**Finding max price of house in every SUBURB**

```python
plt.figure(figsize=(15,10))
vis = avg_price_by_SUBURB.sort_values(by='PRICE',ascending=False)
sns.barplot(x=vis.SUBURB[:25],y=vis.PRICE[:25])
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(15,10))
vis = avg_price_by_SUBURB.sort_values(by='PRICE',ascending=False)
sns.barplot(x=vis.SUBURB[250:],y=vis.PRICE[250:])
plt.xticks(rotation=45)
plt.show()
```

```python
# Price
fig, ax = plt.subplots(1,2, figsize=(16,4))
sns.histplot(x=data_editing_copy['PRICE'], data=data_editing_copy, kde=True, ax=ax[0])
sns.boxplot(x=data_editing_copy['PRICE'], data=data_editing_copy, ax=ax[1])
fig.show()
```

```python
# BEDROOMS
fig, ax = plt.subplots(1,2, figsize=(16,4))
sns.histplot(x=data_editing_copy['BEDROOMS'], data=data_editing_copy, bins=10, ax=ax[0])
sns.boxplot(x=data_editing_copy['BEDROOMS'], data=data_editing_copy, ax=ax[1])
fig.show()
```
