<h1><center>Cars in Russia </center></h1>

**My data was collected at the end of 2020 from a popular car service in Russia.**

**With this data, we can analyze the machines according to their characteristics. For example, prices and model of a car, mileage of a car, maximum speed of a car, volume of a car, fuel of a car. We can also find out the most expensive car. We can find out the ratio of the price of the car for the year of its production.**

**Link to data => https://www.kaggle.com/beaver68/cars-dataset-in-russia**

```python
import pandas as pd
```

```python
my_data = pd.read_csv("cars_dataset.csv")
my_data.head(5)
```

```python
my_data.shape
```

**Lets see column names**

```python
copy_data = my_data.copy()
list(copy_data.columns)
```

**As we see, I have 56 columns. I decided drop some columns. And work with columns, which more interesting. But, every column is important.**

```python
data = copy_data[['mark', 'Box', 'Drive_unit', 'Seats', 'Country', 'Engine', 'Volume',
 'Color', 'Mileage', 'Price', 'Year', 'Car_type', 'Model', 'Cylinders', 'Maximum_speed']]
data.head(10)
```

```python
data.sample(10)
```

```python
# check for NaN objects
data.isnull().sum()
```

```python
# Show mark names
data['mark'].drop_duplicates().to_list()
```

**Lets look at how many cars there were for each brand**

```python
data.mark.value_counts()
```

**Lets see how much revenue each brand**

```python
# I find max and min price each car marks
data.loc[data.Price.notnull()].groupby('mark').Price.agg([sum, min, max ,'count'])
```

**Let's look at 'Maserati' only**

```python
data.loc[data.mark == 'Maserati '].sort_values(['Year','Model'],ascending=False)
```

**Some cars that were made afer 2014**

```python
data.loc[data.Year > 2014].sort_values(by='Year')
```

**Some cars that run on hybrid (gasoline, diesel), and sorting by Price**

```python
data_hybrid_car = data.groupby('Engine').get_group('hybrid')
data_hybrid_car.sort_values(by='Price', ascending=False)
```

**Lets take Subaru model 'Forester', and show its 'Price', 'Mileage', 'Year', 'Engine', 'Color'**

```python
data_Subaru = data.loc[(data.mark == 'Subaru ') & (data.Model=='Forester '), ['Price', 'Mileage', 'Year', 'Engine', 'Color']]
data_Subaru 
```

```python
data_Subaru_ = data.loc[(data.mark == 'Subaru ') & (data.Model=='Forester ')]
data_Subaru_ = data_Subaru_[['mark', 'Model']]
data_Subaru_
```

**Concat two dataframe**

```python
data_Subaru_r = pd.concat([data_Subaru_, data_Subaru], axis=1)
data_Subaru_r.sort_values(by='Price', ascending=True)
```

**Max and min Price car**

```python
data.groupby('Price').get_group(data['Price'].max())
```

```python
data.groupby('Price').get_group(data['Price'].min())
```

**Lets see Mileage of Toyota (max, min, mean)**

```python
Mileage_data_Toyota = data.groupby('mark').get_group('Toyota ')
Mileage_data_Toyota
```

```python
print('max Mileage = ', Mileage_data_Toyota.Mileage.max())
print('min Mileage = ',Mileage_data_Toyota.Mileage.min())
print('mean Mileage = ',Mileage_data_Toyota.Mileage.mean())
```

## Visualization 

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
sns.set(style='whitegrid')
```

**Lets use last data (Toyota)**

```python
plt.rcParams["figure.figsize"] = (20,10)
fig = plt.figure()
ax1 = fig.add_subplot()
ax1.set_ylabel('Mileage')
ax1.set_xlabel('Price')
ax1.set_title('Mileage vs. Price of Toyota')
ax1.scatter(Mileage_data_Toyota.Price, Mileage_data_Toyota.Mileage)
```

**Visualization by Cylinders in car**

```python
sns.countplot(data['Cylinders'])
plt.title('Number of Cylinders')
```

**Plot by seats in every car**

```python
sns.countplot(data['Seats'], color='green')
plt.title('Seats in car')
```

**Counting car marks, and ploting**

```python
mark = pd.DataFrame(data['mark'])
mark_by_count = mark.groupby(['mark']).size().reset_index(name='Count')
mark_by_count
```

```python
plt.figure(figsize=(15,10))
mark_by_count = mark_by_count.sort_values(by='Count',ascending=False)
sns.barplot(x=mark_by_count.mark[:35],y=mark_by_count.Count[:35])
plt.xticks(rotation=90)
plt.show()
```

**Counting by Country**

```python
Country_ = pd.DataFrame(data['Country'])
Country_by_count = Country_.groupby(['Country']).size().reset_index(name='Count')
Country_by_count
```

```python
plt.figure(figsize=(15,10))
Country_by_count = Country_by_count.sort_values(by='Count',ascending=False)
sns.barplot(x=Country_by_count.Country[:35],y=Country_by_count.Count[:35])
plt.xticks(rotation=90)
plt.show()
```

**Ploting Dodge car price by model**

```python
data_Dodge = data.groupby('mark').get_group('Dodge ').reset_index(drop=True)
data_Dodge
```

```python
plt.rcParams["figure.figsize"] = (20,10)
plt.scatter(x=data_Dodge['Model'], y=data_Dodge['Price'])
plt.title('Price x Model Dodge car')
plt.xlabel('Model')
plt.ylabel('Price')
plt.plot()
```
